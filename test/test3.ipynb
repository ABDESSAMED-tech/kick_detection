```python
def sig_var(df,target):
   df = df.reset_index()
   df['variation_'+target] = 0
   for index, row in df.iterrows():
       if index == 0:
           prev_row = row
           df['variation_'+target][index] = df[target][index]
           continue
       df['variation_'+target][index] = row[target] - prev_row[target]
       prev_row = row
   return df
```
